```python
# Create a list
products = ["tshirt", "mug", "hat", "book", "keychain"]

# Create a dictionary
inventory = {}

# Add user input to dictionary
for product in products:
    quantity = int(input(f"Enter the quantity of {product}: "))
    inventory[product] = quantity

# Create a set for customer_orders
customer_orders = set()

# Prompt the user to enter the name of a product that a customer wants to order
purchased = input("Please enter the name of a product you would like to buy: ")
if purchased in products:
  customer_orders.add(purchased)  # Add the product to customer_orders set

# Ask the user if they want to buy another product - continue until they say no
answer = input("Would you like to buy another product? (yes/no) ")
while answer.lower() == "yes":
    purchased = input("What else would you like to buy? ")
    customer_orders.add(purchased)
    answer = input("Would you like to buy another product? (yes/no) ")

# Display the customer's order
print("We have received your order for", customer_orders)
print("Thank you for your order!")
```
